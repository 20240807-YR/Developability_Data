```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ART_DIR = Path("../artifacts/core11")

decision = pd.read_csv(ART_DIR / "decision_log.csv")
state = pd.read_csv(ART_DIR / "state_log.csv")
fallback = pd.read_csv(ART_DIR / "fallback_log.csv")
audit = pd.read_csv(ART_DIR / "audit_log.csv")

decision.head(), state.head()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(state["step"], state["SoMS"])
plt.xlabel("step")
plt.ylabel("SoMS")
plt.title("SoMS Trajectory")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(state["step"], state["toggle_rate"], label="toggle_rate")
plt.plot(state["step"], state["drift"], label="drift")
plt.legend()
plt.title("Toggle & Drift")
plt.show()
```

```python
alloc_codes = (
    state["current_allocation"]
    .astype("category")
    .cat.codes
)

plt.figure(figsize=(10, 3))
plt.step(state["step"], alloc_codes, where="post")
plt.yticks(
    range(len(state["current_allocation"].unique())),
    state["current_allocation"].astype("category").cat.categories,
)
plt.xlabel("step")
plt.ylabel("allocation")
plt.title("Allocation Timeline")
plt.show()
```
